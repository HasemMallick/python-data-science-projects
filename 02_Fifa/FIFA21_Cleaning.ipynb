### 11. Contract column

```python
df["Contract"].unique() # inconsistent in the data
```

```python
# finding all the rows with On loan or Free
for index, row in df.iterrows():
    if "On Loan" in row["Contract"] or "Free" in row['Contract']:
        print(row["Contract"])
```

```python
# Filter rows where 'Contract' contains the substring 'On Loan', handling NaNs
filtered_df = df[df['Contract'].str.contains('On Loan', na=False)]
filtered_df
```

```python
# Remove rows where 'Contract' contains 'On Loan' or 'Free'
df = df[~df['Contract'].str.contains('On Loan|Free', case=False, na=False)]

# Print the filtered DataFrame
df["Contract"]
```

```python
df["Contract"]
```

```python
# Split the 'Contract' column into two new columns
df[['start_contract', 'end_contract']] = df['Contract'].str.split(' ~ ', expand=True)

# Display the resulting DataFrame
print(df)
```

```python
df['start_contract']
```

```python
df["end_contract"]
```

```python
df["start_contract"] = df["start_contract"].astype(int)
df["end_contract"] = df["end_contract"].astype(int)
```

```python
df["end_contract"]
```

```python
# Calculate the duration and store it in a new column 'remaining_time'
df['remaining_time'] = df['end_contract'] - df['start_contract']

# Display the resulting DataFrame
df['remaining_time'].unique()
```

```python
df[["start_contract", "end_contract", "remaining_time"]]
```

```python
# add contract status column 

# Define the function to determine the contract status
def check_status(st):
    if "Free" in st:
        return "Free"
    elif "Loan" in st:
        return "Loan"
    else:
        return "Contract"

# Apply the function to the 'Contract' column
df['Contract_Status'] = df['Contract'].apply(check_status)

# Display the resulting DataFrame
print(df)
```

### 13. Height

```python
df["Height"] = df["Height"].str.rstrip("cm")
df["Height"].unique()
```

```python
df["Height"] = df["Height"].str.replace("\'", "*")
df["Height"].unique()
```

```python
df["Height"] = df["Height"].str.replace(pat='"', repl="")
df["Height"].unique()
```

```python

# Function to convert height from pattern format to cm
def height_pattern_to_cm(height):
    if "*" in height:
        # Split by '*' and strip any extra whitespace
        parts = height.strip().split('*')
        
        # Ensure there are exactly two parts (feet and inches)
        if len(parts) != 2:
            raise ValueError("Invalid height pattern format. Expected format: 'feet*inches'.")
        
        feet = int(parts[0])
        inches = int(parts[1])
        
        total_inches = feet * 12 + inches
        cm = int(total_inches * 2.54)
        return cm
    else:
       return height

# Apply the function to the DataFrame column
test = df['Height'].apply(height_pattern_to_cm)

# Display the updated DataFrame
test.unique()
```

```python
df["Height"] = df["Height"].apply(height_pattern_to_cm)
df["Height"].unique()
```

```python
# replace the column name 
df = df.rename(columns={"Height": "Height(cm)"})
df.columns
```

```python
def format_weight(x):
    if "kg" in x:
        return x.rstrip("kg")
    elif "lbs" in x:
        x = x.rstrip("lbs")
        x = int(x)
        return int(x * 0.45)
```
